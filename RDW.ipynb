```python
import numpy as np
import pandas as pd
import requests as req
from IPython.display import display

pd.set_option("max_columns", None) # show all cols
pd.set_option('max_colwidth', None) # show full width of showing cols
pd.set_option("expand_frame_repr", False) # print cols side by side as it's supposed to be
```

```python
# https://opendata.rdw.nl/Voertuigen/Open-Data-RDW-Gekentekende_voertuigen/m9d7-ebf2
rdw_url_gv = "https://opendata.rdw.nl/resource/m9d7-ebf2.json"
req_rdw_gv = req.get(rdw_url_gv)
pd_rdw_gv = pd.json_normalize(req_rdw_gv.json())
display(pd_rdw_gv.head())
```

```python
# https://opendata.rdw.nl/Voertuigen/Open-Data-RDW-Gekentekende_voertuigen_brandstof/8ys7-d773
rdw_url_gvb = "https://opendata.rdw.nl/resource/8ys7-d773.json"
req_rdw_gvb = req.get(rdw_url_gvb)
pd_rdw_gvb = pd.json_normalize(req_rdw_gvb.json())
display(pd_rdw_gvb)
```

```python
# vind de plug-in hybrid elektrische voertuigen OVC-HEV
display(pd_rdw_gvb.groupby(['klasse_hybride_elektrisch_voertuig']).count())
```

```python
# https://opendata.rdw.nl/Voertuigen/Elektrische-voertuigen/w4rt-e856
rdw_url_ev = "https://opendata.rdw.nl/resource/w4rt-e856.json"
req_rdw_ev = req.get(rdw_url_ev)
pd_rdw_ev = pd.json_normalize(req_rdw_ev.json())
display(pd_rdw_ev)
```

```python
# welk automerk is hot?
display(pd_rdw_ev.groupby(['merk']).count())
```

```python
# welke handelsbenamingen of modellen zijn hot?
display(pd_rdw_ev.groupby(['merk', 'handelsbenaming']).count())
```

```python
# verwijderd NaN van de kolom catalogusprijs
pd_rdw_ev.dropna(subset=['catalogusprijs'], inplace=True)
display(pd_rdw_ev)
```

```python
# zet string om naar integer
pd_rdw_ev['catalogusprijs'] = pd_rdw_ev['catalogusprijs'].astype(int)
```

```python
# gemiddelde aanschafprijs voor elektrische auto's
pd_rdw_ev['catalogusprijs'].mean()
```
